```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
full = pd.concat([train.drop('SalePrice', axis=1), test])
```

```python
missing = full.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
full['LotFrontage'] = full['LotFrontage'].fillna(full['LotFrontage'].median())
full['GarageType'] = full['GarageType'].fillna('None')
```

```python
from sklearn.preprocessing import LabelEncoder
for col in full.select_dtypes(include='object').columns:
    full[col] = full[col].fillna('None')
    full[col] = LabelEncoder().fit_transform(full[col])
```

```python
full['TotalSF'] = full['TotalBsmtSF'] + full['1stFlrSF'] + full['2ndFlrSF']
full['TotalBath'] = full['FullBath'] + (0.5 * full['HalfBath'])
full['HasPool'] = full['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
```

```python
from scipy.stats import skew
numeric_feats = full.dtypes[full.dtypes != "object"].index
skewed_feats = full[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewed = skewed_feats[abs(skewed_feats) > 0.75].index
full[skewed] = np.log1p(full[skewed])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
full[numeric_feats] = scaler.fit_transform(full[numeric_feats])
```

```python
X_train = full[:train.shape[0]]
X_test = full[train.shape[0]:]
y_train = train['SalePrice']
```
