# **Load** **the** **Iris** **Dataset**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load iris dataset from seaborn
df = sns.load_dataset('iris')
df.head()
```

# **Basic** **Exploration**

```python
# Dataset info
df.info()

# Summary stats
df.describe()

# Class distribution
df['species'].value_counts()
```

# **Pairplot**

```python
sns.pairplot(df, hue='species', height=2.5)
plt.suptitle("Pairwise Feature Relationships by Species", y=1.02)
plt.show()
```

# **Boxplot of Petal Length by Species**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_length', data=df, palette='pastel')
plt.title("Petal Length by Species")
plt.show()
```

# **Violin Plot of Sepal Width by Species**

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='sepal_width', data=df, palette='muted')
plt.title("Sepal Width Distribution by Species")
plt.show()
```

# **Correlation Heatmap**

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

# **Scatter** **Plot**

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', style='species', s=100,data = df)
plt.title("Petal Length vs Width by Species")
plt.show()
```
