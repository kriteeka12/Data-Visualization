# **Load** **Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Look at the first few rows
df.head()
```

# **Survival Distribution**

```python
plt.figure(figsize=(8, 5))
x = sns.countplot(x='survived', data=df,color = '#7396ba')
x.set_title("Number of Survivors vs Non-Survivors",fontdict={'family': 'serif','color':'black' ,'size':20})
plt.xlabel("Survived",fontdict = {'family':'serif','color':'black','size':15})
plt.ylabel("Count",fontdict = {'family':'serif','color':'black','size':15})
plt.show()
```

# **Gender Distribution**

```python
plt.figure(figsize=(8,5))
x = sns.countplot(x='sex', data=df,color = '#7396ba')
plt.title("Number of Passengers by Gender",fontdict={'family': 'serif','color':'black' ,'size':20})
plt.xlabel("Gender",fontdict = {'family':'serif','color':'black','size':15})
plt.ylabel("Count",fontdict = {'family':'serif','color':'black','size':15})
plt.show()
```

# **Survival by Gender**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='sex', hue='survived', data=df,palette={0: '#bbe4fa', 1: '#7396ba'})
plt.title("Survival Count by Gender",fontdict={'family': 'serif','weight':'bold','color':'black' ,'size':20})
plt.show()
```

# **Survival by Class**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='pclass', hue='survived', data=df,palette={0: '#bbe4fa', 1: '#7396ba'})
plt.title("Survival by Passenger Class",fontdict={'family': 'serif','weight':'bold','color':'black' ,'size':20})
plt.show()
```

# **Age Distribution**

```python
plt.figure(figsize=(10,5))
sns.histplot(df['age'], kde=True, color='#7396ba')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

# **Age vs Survival**

```python
plt.figure(figsize=(10,5))
sns.histplot(data =df,x = 'age',hue = 'survived', kde=True,multiple='dodge', palette={0: '#bbe4fa', 1: '#7396ba'})
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
```

# **Fare Distribution by Class**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='pclass', y='fare', data=df, palette='pastel')
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

# **Fare Distribution by Survival**

```python
plt.figure(figsize=(8,5))
sns.boxplot(hue='survived', y='fare', data=df,legend = False, palette={0: '#FF6B6B', 1: '#1DD1A1'})
plt.title("Fare Distribution by Survival Status")
plt.xlabel("Survival (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
```
